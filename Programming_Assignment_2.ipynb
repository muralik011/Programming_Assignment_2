4. Write a Python program to solve quadratic equation?

```python
import numpy as np
a = float(input('Enter a: '))
b = float(input('Enter b: '))
c = float(input('Enter c: '))
print(f'Equation: {a}x\N{SUPERSCRIPT TWO} + {b}x + {c}')


f2 = np.sqrt((b ** 2) - (4 * a * c))
x1 = ((-1 * b) + f2) / (2 * a)
x2 = ((-1 * b) - f2) / (2 * a)

print(x1, x2)
```

5. Write a Python program to swap two variables without temp variable?

```python
a, b = 1, 2
print(f'a: {a}, b: {b}')
[a, b] = reversed([a, b])
print(f'a: {a}, b: {b}')
```
